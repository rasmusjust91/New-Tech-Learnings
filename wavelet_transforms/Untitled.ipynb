```python
def Haar(t):
    if 0 <= t < 1/2:
        return 1
    elif 1/2 <= t < 1:
        return -1
    else:
        return 0
```

```python
from math import sin, cos
from random import uniform

def Wave(t):
    return 0.4*sin(t)+0.7*cos(3*t)+2+uniform(0,0.5)
```

```python
import numpy as np
```

```python
uniform(0,1)
```

```python
t = np.linspace(-10, 10, 1000)
```

```python
y_t = np.array(list(map(Wave, t)))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(t,y_t)
```

```python
haar_t = np.array(list(map(Haar, t)))
```

```python
plt.plot(t,haar_t)
```
